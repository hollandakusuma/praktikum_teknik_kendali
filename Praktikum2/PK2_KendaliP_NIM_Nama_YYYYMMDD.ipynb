**5. Pengaturan Parameter Dasar**

```python
setpoint = 40          # cm
batas_persen = 0.05    # ±5% settling band
sampling_time = 10     # ms

kp_rendah=0.1
kp_sedang=0.5
kp_tinggi=1
```

**6. Fungsi Perhitungan Parameter**


*   Rise Time
*   Overshoot
*   Settling Time
*   SSE (Steady-State Error)
*   STD Steady-State (Standar Deviasi Steady-State)
*   Gangguan


```python
# FUNGSI PERHITUNGAN PARAMETER

def hitung_rise_time(df):
    threshold = 0.9 * setpoint
    kondisi = df["jarak_cm"] <= threshold
    if kondisi.any():
        return df.loc[kondisi, "waktu_ms"].iloc[0]
    return None

def hitung_overshoot(df):
    ymax = df["jarak_cm"].max()
    return ((ymax - setpoint) / setpoint) * 100

def hitung_settling_time(df):
    batas_atas = setpoint * (1 + batas_persen)
    batas_bawah = setpoint * (1 - batas_persen)

    for i in range(len(df)):
        data_setelah = df.iloc[i:]
        if ((data_setelah["jarak_cm"] >= batas_bawah) &
            (data_setelah["jarak_cm"] <= batas_atas)).all():
            return df["waktu_ms"].iloc[i]
    return None

def hitung_sse(df):
    data_akhir = df[df["waktu_ms"] >= df["waktu_ms"].max() - 2000]
    return abs(data_akhir["jarak_cm"].mean() - setpoint)

def hitung_std_steady(df):
    data_akhir = df[df["waktu_ms"] >= df["waktu_ms"].max() - 2000]
    return data_akhir["jarak_cm"].std()

def hitung_gangguan(df):
    deviasi_maks = abs(df["jarak_cm"] - setpoint).max()

    batas_atas = setpoint * (1 + batas_persen)
    batas_bawah = setpoint * (1 - batas_persen)

    for i in range(len(df)):
        if (df["jarak_cm"].iloc[i] >= batas_bawah and
            df["jarak_cm"].iloc[i] <= batas_atas):
            return deviasi_maks, df["waktu_ms"].iloc[i]

    return deviasi_maks, None
```

**7. Perhitungan Parameter Tiap Eksperimen**

```python
#Eksperimen 1

#Eksperimen 2

#Eksperimen 3
rise3 = hitung_rise_time(df3)
over3 = hitung_overshoot(df3)
settle3 = hitung_settling_time(df3)
sse3 = hitung_sse(df3)
std3 = hitung_std_steady(df3)
dev3, recover3 = hitung_gangguan(df3)

#Eksperimen 4

#Eksperimen 5

#Eksperimen 6
rise6 = hitung_rise_time(df6)
over6 = hitung_overshoot(df6)
settle6 = hitung_settling_time(df6)
sse6 = hitung_sse(df6)
std6 = hitung_std_steady(df6)
dev6, recover6 = hitung_gangguan(df6)
```

```python
# PLOT GRAFIK RESPON SISTEM PADA EKSPERIMEN 3
plt.figure(figsize=(12, 7))

waktu = df3["waktu_ms"]
jarak = df3["jarak_cm"]

rise_time = rise3
overshoot = over3
settling_time = settle3
sse = sse3

ymax = jarak.max()
batas_atas = setpoint * (1 + batas_persen)
batas_bawah = setpoint * (1 - batas_persen)
threshold_rise = 0.9 * setpoint

plt.plot(waktu, jarak, label="Eksperimen 3")
plt.axhline(setpoint, color='red', linestyle='--', label=f"Setpoint ({setpoint} cm)")
plt.axhline(batas_atas, color='gray', linestyle=':', label=f"Settling Band ({batas_persen*100}%)")
plt.axhline(batas_bawah, color='gray', linestyle=':')

# Tandai Rise Time
if rise_time is not None:
    plt.axvline(x=rise_time, color='green', linestyle='-.', label=f"Rise Time = {rise_time:.0f} ms")
    plt.text(rise_time, threshold_rise, f"  Rise Time = {rise_time:.0f} ms", verticalalignment='bottom')

# Tandai Settling Time
if settling_time is not None:
    plt.axvline(x=settling_time, color='purple', linestyle='-.', label=f"Settling Time = {settling_time:.0f} ms")
    plt.text(settling_time, setpoint + 5, f"  Settling = {settling_time:.0f} ms", verticalalignment='top', fontsize=14)

# Tandai Overshoot
if overshoot > 0:
    plt.scatter(waktu.iloc[jarak.idxmax()], ymax, color='orange', zorder=5)
    plt.text(waktu.iloc[jarak.idxmax()], ymax, f"  Overshoot = {overshoot:.2f}%", verticalalignment='bottom')

# Kotak informasi parameter
textstr = '\n'.join((
    f'Rise Time : {rise_time} ms',
    f'Overshoot : {overshoot:.2f} %',
    f'Settling  : {settling_time} ms',
    f'SSE       : {sse:.3f} cm'
))

plt.gca().text(0.5, 0.98, textstr,
               transform=plt.gca().transAxes, fontsize=12,
               verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))

plt.title("Respon Sistem - Eksperimen 3", fontsize=16)
plt.xlabel("Waktu (ms)", fontsize=14)
plt.ylabel("Jarak (cm)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```
