**8. Implementasi Moving Average Filter**

Fungsi ini menggunakan metode rolling() dari pandas.

📌 Membuat Fungsi Moving Average

```python
def moving_average(data, window_size):
    return data.rolling(window=window_size).mean()
```

📌 Menambahkan Variasi Window

```python
window_list = [3, 5, 10, 20]

for w in window_list:
    df1[f"MA_{w}"] = moving_average(df1["jarak_cm"], w)
    #df2[f"MA_{w}"] = moving_average(df2["jarak_cm"], w)
    #df3[f"MA_{w}"] = moving_average(df3["jarak_cm"], w)
    #df4[f"MA_{w}"] = moving_average(df4["jarak_cm"], w)
```

**13. Evaluasi Kuantitatif Performa Moving Average**

**A. Noise Reduction Ratio (NRR)**

Ukuran seberapa efektif filter mereduksi noise.
$$
NRR =  \frac{\sigma_{raw}-\sigma_{Moving Averaage}}{\sigma_{raw}}\times 100 \%
$$

Semakin besar → semakin baik dalam mereduksi noise.

```python
def noise_reduction_ratio(raw, filtered):
    std_raw = raw.std()
    std_filtered = filtered.std()
    nrr = (std_raw - std_filtered) / std_raw * 100
    return std_raw, std_filtered, nrr
```

```python
#Eksperimen 1

from IPython.display import HTML
results = []
for w in window_list:
    std_raw, std_ma, nrr = noise_reduction_ratio(df1["jarak_cm"], df1[f"MA_{w}"])
    results.append({"Window Size": f"MA_{w}", "STD Raw": std_raw, "STD Filtered": std_ma, "NRR (%)": nrr})

df_results = pd.DataFrame(results)

print("HASIL EKSPERIMEN 1")
# Convert DataFrame to HTML and display with borders
html_table = df_results.to_html(index=False, formatters={"STD Raw": "{:.4f}".format, "STD Filtered": "{:.4f}".format, "NRR (%)": "{:.2f}".format}, border=1)
display(HTML(html_table))

#Eksperimen 2

#Eksperimen 3

```

**B. Root Mean Square Error (RMSE) terhadap Nilai Referensi**

Karena eksperimen 1 dan 2 memiliki jarak tetap (50 cm dan 30 cm), kita bisa ukur deviasi terhadap nilai ideal.

$$
RMSE = \sqrt{\frac{1}{N}\sum{(y-y_{referensi})^2}}
$$

```python
import numpy as np

def rmse(data, reference):
    return np.sqrt(np.mean((data - reference)**2))
```

```python
#EKSPERIMEN 1
reference = 50

from IPython.display import HTML
results_rmse = []

# Get RMSE for raw data once
rmse_raw = rmse(df1["jarak_cm"], reference)

for w in window_list:
    rmse_ma = rmse(df1[f"MA_{w}"], reference)
    results_rmse.append({"Window Size": f"MA_{w}", "RMSE Raw": rmse_raw, "RMSE Filtered": rmse_ma})

df_rmse_results = pd.DataFrame(results_rmse)

print("HASIL RMSE EKSPERIMEN 1")
# Convert DataFrame to HTML and display with borders
html_table_rmse = df_rmse_results.to_html(index=False, formatters={"RMSE Raw": "{:.4f}".format, "RMSE Filtered": "{:.4f}".format}, border=1)
display(HTML(html_table_rmse))

#EKSPERIMEN 2
```

**C. Delay Time (Time Lag) - Eksperimen 3**

Untuk eksperimen step (Eksperimen 3), kita ukur delay sebagai: Waktu saat sinyal mencapai **90%** perubahan.

Misalnya

**step dari 50 ke 30 (Delay Turun):**

Target 90%:
$$
y_{90\%}=50-0.9(50-30)=32
$$


**step dari 30 ke 50: (Delay Naik):**

Target 90%:
$$
y_{90\%}=30+0.9(50-30)=48
$$







**Fungsi sederhana mencari waktu crossing**

```python
def hitung_delay_naik(data, waktu, nilai_awal, nilai_akhir):
    delta = nilai_akhir - nilai_awal
    threshold = nilai_awal + 0.9 * delta  # 90% perubahan

    t0 = None
    t90 = None

    # Cari t0 (awal perubahan naik)
    for i in range(1, len(data)):
        if data.iloc[i] > nilai_awal + 0.05 * delta:  # mulai meninggalkan steady-state
            t0 = waktu.iloc[i]
            break

    # Cari t90
    for i in range(len(data)):
        if data.iloc[i] >= threshold:
            t90 = waktu.iloc[i]
            break

    if t0 is not None and t90 is not None:
        return t0, t90, t90 - t0
    else:
        return t0, t90, None
```

```python
nilai_awal = 30
nilai_akhir = 50

#Data Eksperimen
t0_raw, t90_raw, delay_raw = hitung_delay_naik(
    df1["jarak_cm"], df1["waktu_ms"], nilai_awal, nilai_akhir
)

#Moving Average 3

#Moving Average 5

#Moving Average 10

#Moving Average 20
t0_ma20, t90_ma20, delay_ma20 = hitung_delay_naik(
    df3["MA_20"], df3["waktu_ms"], nilai_awal, nilai_akhir
)

print("RAW  : t0 =", t0_raw, " t90 =", t90_raw, " delay =", delay_raw)
print("MA20 : t0 =", t0_ma20, " t90 =", t90_ma20, " delay =", delay_ma20)
```

```python
def hitung_delay_turun(data, waktu, nilai_awal, nilai_akhir):
    delta = nilai_awal - nilai_akhir
    threshold = nilai_awal - 0.9 * delta  # 90% perubahan

    t0 = None
    t90 = None

    # Cari t0 (awal perubahan)
    for i in range(1, len(data)):
        if data.iloc[i] < nilai_awal - 0.05 * delta:  # mulai meninggalkan steady-state
            t0 = waktu.iloc[i]
            break

    # Cari t90
    for i in range(len(data)):
        if data.iloc[i] <= threshold:
            t90 = waktu.iloc[i]
            break

    if t0 is not None and t90 is not None:
        return t0, t90, t90 - t0
    else:
        return t0, t90, None
```

```python
nilai_awal = 50
nilai_akhir = 30

#Data Eksperimen
t0_raw, t90_raw, delay_raw = hitung_delay_turun(
    df3["jarak_cm"], df3["waktu_ms"], nilai_awal, nilai_akhir
)

#Moving Average 3

#Moving Average 5

#Moving Average 10

#Moving Average 20
t0_ma20, t90_ma20, delay_ma20 = hitung_delay_turun(
    df3["MA_20"], df3["waktu_ms"], nilai_awal, nilai_akhir
)

print("RAW  : t0 =", t0_raw, " t90 =", t90_raw, " delay =", delay_raw)
print("MA20 : t0 =", t0_ma20, " t90 =", t90_ma20, " delay =", delay_ma20)
```
